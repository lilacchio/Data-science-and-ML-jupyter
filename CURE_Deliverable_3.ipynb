# CURE 3

Imports

```python
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
```

## Task 2

Read in the data from the CSV

```python
horse_data = pd.read_csv("horse.csv")
horse_data = horse_data.drop(columns=['surgery', 'age', 'outcome', 'lived_numerical', 
                                      'died_numerical.1', 'surgery_numerical', 'no_surgery_numerical'])
horse_data.head(100)
```

Replace all the **Nan**'s with median values of the dataset

```python
imputer = SimpleImputer(strategy="median")
imputer.fit(horse_data)
X = imputer.transform(horse_data)
horse_data = pd.DataFrame(X, columns=horse_data.columns, index=horse_data.index)
```

Display the data

```python
horse_data.head()
```

## Task 3

Scatterplots - Combined

```python
lived_with_surgery = horse_data[horse_data.lived_with_surgery_numerical == 1]
lived_without_surgery = horse_data[horse_data.lived_without_surgery_numerical == 1]
died = horse_data[horse_data.died_numerical == 1]

fig, axis = plt.subplots(figsize=(9,3))
scatter1 = plt.subplot2grid(shape=(1,3), loc = (0, 0))
scatter2 = plt.subplot2grid(shape=(1,3), loc = (0, 1))
scatter3 = plt.subplot2grid(shape=(1,3), loc = (0, 2))
plt.tight_layout()

scatter1.scatter(lived_with_surgery["pulse"], lived_with_surgery["rectal_temp"], label="lived with surgery", edgecolor="none", fc="red")
scatter1.scatter(lived_without_surgery["pulse"], lived_without_surgery["rectal_temp"], label="lived without surgery", edgecolor="none", fc="black")
scatter1.scatter(died["pulse"], died["rectal_temp"], label="died", edgecolor="none", fc="blue")
scatter1.set_xlabel("pulse")
scatter1.set_ylabel("temperature")
scatter1.legend()

scatter2.scatter(lived_with_surgery["pulse"], lived_with_surgery["respiratory_rate"], label="lived with surgery", edgecolor="none", fc="red")
scatter2.scatter(lived_without_surgery["pulse"], lived_without_surgery["respiratory_rate"], label="lived without surgery", edgecolor="none", fc="black")
scatter2.scatter(died["pulse"], died["respiratory_rate"], label="died", edgecolor="none", fc="blue")
scatter2.set_xlabel("pulse")
scatter2.set_ylabel("respiratory_rate")
scatter2.legend()

scatter3.scatter(lived_with_surgery["respiratory_rate"], lived_with_surgery["rectal_temp"], label="lived with surgery", edgecolor="none", fc="red")
scatter3.scatter(lived_without_surgery["respiratory_rate"], lived_without_surgery["rectal_temp"], label="lived without surgery", edgecolor="none", fc="black")
scatter3.scatter(died["respiratory_rate"], died["rectal_temp"], label="died", edgecolor="none", fc="blue")
scatter3.set_xlabel("respiratory_rate")
scatter3.set_ylabel("temperature")
scatter3.legend()
```
